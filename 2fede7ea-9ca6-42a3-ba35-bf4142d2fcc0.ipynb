```python
# since there are many individual years, I will categorize them into significant periods and filter by that information
# chose the years based on an even distribution between the minimum and maximum years present in the dataframe to avoid skewed results
def year_period(year):
    """
    The function returns the period group for games depending on the year of release, using the following rules:
    -'retro' for year <1992
    -'modern' for 1992 <= year < 2004
    -'recent' for year >= 2004
    -'unknown' for missing year values (NaN)
    """
    if year < 1992:
        return 'retro'
    elif year < 2004:
        return 'modern'
    elif year >= 2004:
        return 'recent'
    else:
        return 'unknown'
    
# creating a column in the dataframe that will label which period the game was released in
dfgames['release_period'] = dfgames['year_of_release'].apply(year_period)

dfgames['release_period'].value_counts()
```

Test the following hypotheses:
- Average user ratings of the Xbox One and PC platforms are the same. 
- Average user ratings for the Action and Sports genres are different.

Set the aplha threshold value yourself. 

Explain:
- How you formulated the null and alternative hypotheses
- What significance level you chose to test the hypotheses, and why

Correct interpretation of the Null and Alternative hypotheses

```python
# was a significant amount of missing values in the 'user_score' column so replacing them with the average user score. Keep in mind how this may reduce variability and mask certain trends. 
user_score_mean = recent_data['user_score'].mean()
recent_data.loc[:, 'user_score'] = recent_data['user_score'].fillna(user_score_mean)
```

```python
# Hypothesis 1: average user rating of the Xbox One and PC platforms are the same
# Null hypothesis = The average user ratings of the Xbox One and PC platforms are equal
# Alternative Hypothesis = The average user ratings of the Xbox One and PC platforms are not equal

# filtering for each specific platform and their user scores
xone_scores = recent_data.query("platform=='XOne'")['user_score']
pc_scores = recent_data.query("platform=='PC'")['user_score']

alpha = 0.05 # critical statistical significance
results = st.ttest_ind(xone_scores, pc_scores, equal_var=False)
print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print("We can't reject the null hypothesis")
    
```

During this hypothesis, the alpha value of 0.05 was chosen due to the fact that it represents a 95% level of confidence with our results. It was also chosen because this value is widely recognized as the standard for testing hypotheses.
